```python
import pandas as pd

df = pd.read_csv("../data/strandings_example.csv")
df
```

```python
import matplotlib.pyplot as plt

df.groupby("species")["count"].sum().plot(kind="bar", color="teal")
plt.title("Total strandings by species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt

df.groupby("region")["count"].sum().plot(kind="bar", color="darkorange")
plt.title("Total de strandings por região")
plt.xlabel("Região")
plt.ylabel("Número de strandings")
plt.show()
```

```python
import matplotlib.pyplot as plt

# garantir que a coluna date é datetime
df["date"] = pd.to_datetime(df["date"])

# agrupar por mês e somar
monthly = df.groupby(df["date"].dt.to_period("M"))["count"].sum()

# converter PeriodIndex para datetime para plotar
monthly.index = monthly.index.to_timestamp()

# plot
monthly.plot(kind="line", marker="o", color="purple")
plt.title("Strandings ao longo do tempo")
plt.xlabel("Mês")
plt.ylabel("Total de strandings")
plt.grid(True)
plt.show()
```
